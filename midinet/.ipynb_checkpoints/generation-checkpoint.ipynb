```python
use_gpu = torch.cuda.is_available()
if use_gpu:
    device = 'cuda'
else:
    device = 'cpu'

    pitch_range = 128

def gpu(tensor, gpu=use_gpu):
    if gpu:
        return tensor.cuda()
    else:
        return tensor
batch_size = 288*2
```

```python
prev_data = torch.Tensor([[[[np.random.binomial(2,0.1) for i in range(128)]for j in range(16)] for k in range(1)] for p in range(batch_size)])
prev_data_cpu = prev_data.to(device)

netG = generator(pitch_range).to(device)
netG.load_state_dict(torch.load('models/netG_epoch_999.pth'))
netG.eval()
```

```python
fixed_noise = gpu(torch.randn(batch_size, 100))
gen_output = netG(fixed_noise,prev_data_cpu,batch_size,pitch_range)
gen_output_2 = netG(fixed_noise,gen_output,batch_size,pitch_range)
gen_output=gen_output.cpu().detach().numpy()
gen_output_2=gen_output_2.cpu().detach().numpy()
```

```python
threshold = 0.1

res = []
s = 0
c = 0
for bar in gen_output:
    res.append((bar[0]>=threshold).T)
    c += 1
    s += np.sum(bar[0]>=threshold)
for bar in gen_output_2:
    res.append((bar[0]>=threshold).T)
    c += 1
    s += np.sum(bar[0]>=threshold)
dico = {"Voice 1":res}

print(s/c)
```

```python
hist = pitches_histogram(dico)
t = best_tonality(hist, print_results=True)
transposed = transpose(dico, t)


score_rec = dictionary_to_midi(transposed)
score_rec.show('midi')
```
